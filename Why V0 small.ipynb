```python
from scipy.optimize import curve_fit
def func(x, vm, b,vin):
    return vm*exp(-((x)/b)**2)+vin

para = [curve_fit(func, X, numpy.mean(Vmat40[i]['TT1'], axis=0)) for i in range(8)]
p = [para[i][0] for i in range(8)]
Qf=array([2700,2250,1800,1350,900,450,300,150])*21.2/2700

def func_line(x, a):
    return a*x
p3, cov3 = curve_fit(func_line, Qf**(1./3.),[p[j][0] for j in range(8)])
line = func_line(Qf**(1./3.), p3[0])

fig = plt.figure(figsize=(12,4))

subplot(1,2,1)
plot(Qf**(1./3.),[p[j][0] for j in range(8)], 'bs', label='$v_m$')
plot(Qf**(1./3.),line, 'r-', label='line fit')
#xlim(-0.4,0)
#ylim(-0.5,1.3)
xlabel('$\dot Q_f^{1/3} / kw^{1/3}$',size=20)
ylabel('$  v_m / m/s$',size=20)
legend(loc='lower right', numpoints=1)
grid()

def func_l(x, a):
    return 0*x+a
p4, cov4 = curve_fit(func_l, Qf,np.absolute([p[j][1] for j in range(8)]))
line2 = func_l(Qf, p4[0])

subplot(1,2,2)
plot(Qf,np.absolute([p[j][1] for j in range(8)]), 'bs', label='$b$')
plot(Qf,line2, 'r-', label='line fit')
#xlim(-0.4,0)
ylim(0,0.25)
xlabel('$\dot Q_f / kw$',size=20)
ylabel('$  b / m$',size=20)
legend(loc='lower right', numpoints=1)
grid()

p3,p4
```
