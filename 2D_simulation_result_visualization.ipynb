```python
"""
Read FDS SLCF file to numpy data
Print the data name, and return the data metrix
author: Qize He
email: hqz@utexas.edu
"""
def readslcf(fname,Tstart,Tend,Tstep):
    fid=open(fname,'rb'); 
    fromfile(fid, int8, 4)
    str1=fromfile(fid, 'b', 30)
    str1=''.join(["%c" % number for number in str1.astype(int)])
    fromfile(fid, int8, 4)
    fromfile(fid, int8, 4)
    str2=fromfile(fid, 'b', 30)
    str2=''.join(["%c" % number for number in str2.astype(int)])
    fromfile(fid, int8, 4)
    fromfile(fid, int8, 4)
    str3=fromfile(fid, 'b', 30)
    str3=''.join(["%c" % number for number in str3.astype(int)])
    fromfile(fid, int8, 4)
    fromfile(fid, int8, 4)
    indx=fromfile(fid, int32, 6)
    fromfile(fid, int8, 4)
    Isize = indx[1]-indx[0]+1
    Jsize = indx[3]-indx[2]+1
    Ksize = indx[5]-indx[4]+1
    if Isize == 1: 
        M = Jsize 
        N = Ksize 
    elif Jsize == 1: 
        M = Isize 
        N = Ksize 
    else: 
        M = Isize
        N = Jsize
    T=numpy.zeros((N, M, 1))
    st = 0
    Time=0
    while Time < Tstart:
        fromfile(fid, int8, 4)
        Timer = fromfile(fid, float32,1)
        Time=Timer[0]
        fromfile(fid, int8, 4)
        fromfile(fid, int8, 4)
        T[:,:,st]=fromfile(fid, float32, M*N).reshape((N, M))
        fromfile(fid, int8, 4)
    while Time < Tend:
        st = st + 1 
        fromfile(fid, int8, 4)
        Timer = fromfile(fid, float32,1)
        Time=Timer[0]
        fromfile(fid, int8, 4)
        fromfile(fid, int8, 4)
        T=np.concatenate((T,numpy.zeros((N, M, 1))), axis=2)
        T[:,:,st]=fromfile(fid, float32, M*N).reshape((N, M))
        fromfile(fid, int8, 4)
    print(str1,str2,str3,T.shape)
    fid.close()
    return T
```

```python
def runningMean(x, N):
    y = np.zeros((len(x)))
    for ctr in range(len(x)):
        y[ctr] = np.average(x[ctr:(ctr+N)])
    return y
```
